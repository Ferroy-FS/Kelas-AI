```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = {
    'SKS': [18, 19, 20, 20, 21, 22, 22, 23, 24, 24],
    'IPS': [3.85, 3.70, 3.65, 3.40, 3.30, 3.15, 3.00, 2.85, 2.70, 2.50]
}

df = pd.DataFrame(data)
print(df)
```

```python
X = df[['SKS']]
Y = df[['IPS']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
print("Intercept  (A):", model.intercept_)
print("Koefisien  (B):", model.coef_[0])
```

```python
Y_pred =  model.predict(X_test)

print("Data Testing:")
print(X_test)
print("IPS Asli:", Y_test.values)
print("Prediksi IPS:", Y_pred)
```

```python
print("MAE:", metrics.mean_absolute_error(Y_test, Y_pred))
print("MSE:", metrics.mean_squared_error(Y_test, Y_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

```python
prediksi = model.predict([[10]])
print("Prediksi IPS untuk 10 semester kuliah", prediksi[0])
```

```python
plt.scatter(X, Y, label="Data Asli")
plt.plot(X, model.predict(X), label="Garis Regresi")
plt.xlabel("SKS (Jumlah)")
plt.ylabel("IPS (Nilai)")
plt.title("Regresi Linear: SKS vs IPS")
plt.legend()
plt.show()
```

Model Regresi: Model menghasilkan persamaan $IPS = 7.94 - 0.22(SKS)$, yang menunjukkan adanya hubungan negatif antara jumlah SKS yang diambil dengan IPS yang diraih.

Akurasi: Nilai Error (RMSE) sebesar 0.07 menunjukkan model ini sangat akurat dalam memprediksi data dalam rentang yang ada.

Kesimpulan: Penambahan beban SKS cenderung menurunkan IPS secara bertahap, kemungkinan karena beban belajar yang semakin berat.
