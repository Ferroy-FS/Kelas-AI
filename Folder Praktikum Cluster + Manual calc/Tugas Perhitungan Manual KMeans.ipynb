```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# 1. Buat Dataset
data = {
    'X': [2, 2, 8, 7, 1, 9, 2, 8, 3, 7],
    'Y': [3, 4, 9, 8, 2, 10, 2, 7, 3, 9]
}
df = pd.DataFrame(data)
```

```python
# 2. Jalankan K-Means
kmeans = KMeans(n_clusters=2, random_state=42).fit(df)
df['cluster'] = kmeans.labels_
```

```python
# 3. Hitung Silhouette Score
score = silhouette_score(df[['X', 'Y']], kmeans.labels_)

print("Hasil Cluster:\n", df)
print(f"\nSilhouette Score: {score:.4f}")
```
